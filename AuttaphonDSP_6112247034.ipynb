```python
#lab01

from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display

sin_sig = thinkdsp.SinSignal(freq=7034, amp=34, offset=0)
sin_sig.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#lab02

signal = thinkdsp.TriangleSignal(4000)
duration = signal.period*2500
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
wave = signal.make_wave(duration, framerate=10000)
wave.apodize()
wave.make_audio()

spectrum = wave.make_spectrum()
spectrum.plot()


```

```python
signal = thinkdsp.SquareSignal(34)
duration = signal.period*34
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
segment.plot()
Twave = signal.make_wave(duration, framerate=10000)
Twave.apodize()
Twave.make_audio()




```
